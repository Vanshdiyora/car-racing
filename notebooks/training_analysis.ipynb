# Training Analysis

Analyse PPO training logs: reward curves, loss trends, and convergence diagnostics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({"figure.figsize": (12, 5), "font.size": 12})

CSV_PATH = "../logs/training_logs.csv"
```

```python
df = pd.read_csv(CSV_PATH)
print(f"Loaded {len(df)} rows")
df.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Reward curve
axes[0].plot(df["step"], df["mean_reward"], color="steelblue")
axes[0].set_title("Evaluation Mean Reward")
axes[0].set_xlabel("Training Step")
axes[0].set_ylabel("Reward")
axes[0].grid(True, alpha=0.3)

# Loss curve
axes[1].plot(df["step"], df["loss"], color="coral")
axes[1].set_title("Training Loss")
axes[1].set_xlabel("Training Step")
axes[1].set_ylabel("Loss")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("../outputs/plots/training_curves.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))

for ax, col, title in zip(
    axes,
    ["pg_loss", "vf_loss", "entropy"],
    ["Policy Loss", "Value Loss", "Entropy"],
):
    ax.plot(df["step"], df[col])
    ax.set_title(title)
    ax.set_xlabel("Step")
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Rolling statistics
window = max(1, len(df) // 20)
df["reward_smooth"] = df["mean_reward"].rolling(window, min_periods=1).mean()

plt.figure(figsize=(12, 5))
plt.plot(df["step"], df["mean_reward"], alpha=0.3, label="Raw")
plt.plot(df["step"], df["reward_smooth"], linewidth=2, label=f"Rolling mean (w={window})")
plt.xlabel("Step")
plt.ylabel("Reward")
plt.title("Smoothed Reward Curve")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
print("=== Training Summary ===")
print(f"Total steps:  {df['step'].max():,}")
print(f"Best reward:  {df['mean_reward'].max():.1f}")
print(f"Final reward: {df['mean_reward'].iloc[-1]:.1f}")
print(f"Final loss:   {df['loss'].iloc[-1]:.4f}")
print(f"Wall time:    {df['time'].iloc[-1]:.0f}s")
```
